```python
 #Pivot table 
data = {'Company': ['A', 'A', 'B', 'B', 'C', 'C'],
        'Year': [2020, 2020, 2021, 2021, 2022, 2022],
        'Sales': [100, 200, 150, 250, 300, 400]}
df3 = pd.DataFrame(data)

result = df3.pivot_table(index='Year', columns='Company', values='Sales')


print(result)
```

```python
#Merge two dataframe 

# Create two dataframes
df4 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index=[0, 1, 2, 3])

df5 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                   index=[4, 5, 6, 7])

# Merge the two dataframes on the 'A' column
result = pd.merge(df4, df5, on='A')

print(result)
```

```python
#Apply function to a dataframe 

# Create a dataframe
df6 = pd.DataFrame({'A': [1, 2, 3, 4],
                   'B': [10, 20, 30, 40],
                   'C': [100, 200, 300, 400]})

# Define a function to be applied to the dataframe
def double_column(col):
    return col * 2

# Apply the function to the entire dataframe
result = df6.apply(double_column)

print(result)
```

```python
#Apply entire function to a entire column
import pandas as pd

# Create a dataframe
df01 = pd.DataFrame({'A': [1, 2, 3, 4],
                   'B': ['dog', 'cat', 'bird', 'fish']})

# Define a function to be applied to the column
def upper_case(string):
    return string.upper()

# Apply the function to the 'B' column
result = df01['B'].map(upper_case)

print(result)
```

```python
#lambda: to create anonymous functions that can be used in combination with other functions such as map and apply.

import pandas as pd


df05 = pd.DataFrame({'A': [1, 2, 3, 4],
                   'B': [10, 20, 30, 40],
                   'C': [100, 200, 300, 400]})


result = df05['B'].apply(lambda x: x + 10)

print(result)

```
